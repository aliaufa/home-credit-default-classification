# 1. Introduction

## Background

Risiko kredit default adalah salah satu masalah utama yang dihadapi oleh perusahaan keuangan seperti Home Credit Indonesia. Ketika memberikan pinjaman kepada pelanggan, perusahaan perlu melakukan evaluasi risiko kredit untuk memastikan bahwa pelanggan memiliki kemampuan untuk membayar kembali pinjaman sesuai dengan kesepakatan. Salah satu cara untuk mengatasi risiko kredit default adalah dengan menggunakan model prediksi yang dapat mengidentifikasi calon debitur yang berpotensi mengalami kesulitan dalam membayar pinjaman.

## Problem Statement

Bagaimana mengembangkan model prediksi yang akurat untuk mengklasifikasikan pemilik kredit sebagai pembayar atau default dalam layanan pembiayaan Home Credit Indonesia? Bagaimana cara mengatasi ketidakseimbangan kelas dalam dataset dan memilih model yang optimal dengan metrik evaluasi yang tepat?

Dalam kasus ini, beberapa metrik yang tepat untuk mengevaluasi kinerja model prediksi risiko kredit default adalah:

- ROC-AUC : Metrik ini mengukur kemampuan model dalam membedakan antara kelas positif (default) dan kelas negatif (non-default). ROC-AUC memberikan gambaran keseluruhan tentang kinerja model dengan mempertimbangkan trade-off antara true positive rate (sensitivity) dan false positive rate (1-specificity). Semakin tinggi nilai ROC-AUC, semakin baik model dalam memprediksi risiko kredit default.

- Recall (Sensitivity): Metrik ini mengukur sejauh mana pemilik kredit yang default yang dapat diidentifikasi oleh model. Recall memberikan gambaran tentang proporsi pemilik kredit default yang berhasil diidentifikasi oleh model dari keseluruhan pemilik kredit default yang sebenarnya. Recall penting untuk memahami sejauh mana model dapat mendeteksi risiko kredit default.

## Kolom Kategorikal

```python
def show_plots(data, column, rotate=0):
    '''
    Show countplot and stacked barplot for a categorical column in a DataFrame.

    Parameters:
        data (DataFrame): The input DataFrame containing the data.
        column (str): The name of the categorical column to plot.
        rotate (int): The rotation angle for x-axis labels in degrees (default: 0).

    Returns:
        None
    '''    
    color_palette = {0: '#70CAb0', 1: '#F38091'}

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Countplot
    sns.countplot(data=data, x=column, hue='TARGET', palette=color_palette, ax=axes[0])
    axes[0].set_title(f'Countplot of {column}')
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=rotate)
    axes[0].legend(title='TARGET', labels=['No', 'Default'])

    # Stacked barplot
    stacked_data = data.groupby(column)['TARGET'].value_counts(normalize=True).unstack()
    stacked_data.plot(kind='bar', stacked=True, color=color_palette, ax=axes[1])
    axes[1].set_title(f'Stacked Bar Plot of {column}')
    axes[1].set_xticklabels(stacked_data.index, rotation=rotate)
    axes[1].legend(title='TARGET', labels=['No', 'Default'])

    plt.tight_layout()
    plt.show()
```

```python
# plot categorical columns
for column in cat_cols:
    if application_train[column].nunique() > 20:
        continue
    elif application_train[column].nunique() > 10:
        show_plots(application_train, column, 90)
    elif application_train[column].nunique() <= 3:
        show_plots(application_train, column, 0)
    else:
        show_plots(application_train, column, 45)
```

## Handling missing values
Missing values akan dilakukan imputasi. Metode imputasi yang dilakukan adalah sebagai berikut:

- Kolom numerik distribusi normal : imputasi dengan nilai mean
- Kolom numerik distribusi skewed : imputasi dengan nilai median
- Kolom kategorikal : imputasi dengan nilai modus

Imputasi akan dilakukan dalam pipeline.

## Handling Outlier
Outlier akan diidentifikasi untuk kolom yang terdistribusi normal dengan z-score, dan untuk kolom terdistribusi skewed dengan iqr.

```python
outlier_percentages = {}

# Calculate outliers using z-score for normal_dist columns
zscore_threshold = 3.0
for col in normal_dist:
    zscores = np.abs(stats.zscore(data[col]))
    outliers_percentage = np.sum(zscores > zscore_threshold) / len(data) * 100
    outlier_percentages[col] = outliers_percentage

# Calculate outliers using IQR for skewed_dist columns
iqr_threshold = 1.5
for col in skewed_dist:
    q25 = np.percentile(data[col], 25)
    q75 = np.percentile(data[col], 75)
    iqr = q75 - q25
    lower_bound = q25 - (iqr_threshold * iqr)
    upper_bound = q75 + (iqr_threshold * iqr)
    outliers_percentage = np.sum((data[col] < lower_bound) | (data[col] > upper_bound)) / len(data) * 100
    outlier_percentages[col] = outliers_percentage

# Print the percentages of outliers
for col, percentage in outlier_percentages.items():
    print(f"{col}: {percentage:.2f}%")
```

Terdapat 6 kolom yang memiliki outlier. Jumlah outlier terbilang sedikit dimana kolom dengan jumlah outlier terbanyak `AMT_INCOME_TOTAL` hanya memiliki 4.56% outlier.

Outlier tidak akan dilakukan handling, namun pada bagian scaling akan digunakan metode RobustScaler yang tidak sensitif terhadap outlier.

## Preprocess Pipeline
Pipeline akan melakukan transformasi fitur. Transformasi yang akan dilakukan adalah impute missing values, scaling fitur numerik, dan encoding fitur kategorikal.

```python
# Create pipelines for different types of columns
normal_dist_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())
])

skewed_dist_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

cat_cols_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

cat_num_cols_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent'))
])
# Combine the pipelines
preprocessing_pipeline = ColumnTransformer(transformers=[
    ('normal_dist', normal_dist_pipeline, normal_dist),
    ('skewed_dist', skewed_dist_pipeline, skewed_dist),
    ('cat_cols', cat_cols_pipeline, cat_cols),
    ('cat_num_cols', cat_num_cols_pipeline, cat_num_cols)
])
```

```python
# Fit the preprocessing pipeline on the data
preprocessing_pipeline.fit(X_train)

# Apply preprocessing pipeline on the data and get transformed features
X_train_transformed = preprocessing_pipeline.transform(X_train)
X_test_transformed = preprocessing_pipeline.transform(X_test)

# Get the feature names after applying OneHotEncoder
cat_cols_encoded_names = preprocessing_pipeline.named_transformers_['cat_cols']['encoder'].get_feature_names_out(cat_cols)

# Concatenate the feature names from all columns
feature_names = np.concatenate([normal_dist, skewed_dist, cat_cols_encoded_names, cat_num_cols])

# Create a DataFrame with transformed data and feature names
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=feature_names)
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=feature_names)

# Print the transformed data
display(X_train_transformed_df)
```

## Reduksi Dimensi - PCA
Untuk mengurangi dimensi dan mempercepat proses training akan dilakukan PCA

```python
# define PCA
pca = PCA()
pca.fit(X_train_transformed_df)

# PCA with 0.95 information

cumsum = np.cumsum(pca.explained_variance_ratio_) # cumsum = kulumasi baris atau kolom berikutnya bergantung axis = 0 atau 1
d = np.argmax(cumsum >= 0.95) + 1                 # argmax = returns the indices of the maximum values along an axis.   

plt.figure(figsize=(6,4))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 50, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Kumulatif Explained Var Ratio")
plt.plot([d, d], [0, 0.95], "k:")           # plt. plot([X1,X2], [y1,y2]), ':k' = black
plt.plot([0, d], [0.95, 0.95], "k:")        # plt. plot([X1,X2], [y1,y2]), ':k' = black
plt.plot(d, 0.95, "ko")                     # plot elbow, ko = circle b
plt.grid(True)
plt.show()
```

Untuk mempertahankan 95% informasi dari seluruh fitur akan digunakan reduksi sampai 32 dimensi. PCA akan dimasukan ke dalam pipeline

```python
# Define the final pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessing_pipeline),  # Preprocessing pipeline
    ('pca', PCA(n_components=32))  # PCA step
    ])
```
